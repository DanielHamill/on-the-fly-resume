```python
from jinja2 import Environment, PackageLoader, select_autoescape
```

```python
env = Environment(
    loader=PackageLoader("on_the_fly_resume", "templates"),
    variable_start_string='((*',
    variable_end_string='*))',
    block_start_string='((%',
    block_end_string='%))',
    comment_start_string='((#',
    comment_end_string='#))',
    autoescape=False  # Don't autoescape for LaTeX
)
```

```python
template = env.get_template("template.tex")
```

```python
print(template.render(resumeItems=["hello world", "another item", "and another"]))
```
